```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect('market_data.db')
```

```python
cursor = conn.cursor()
cursor.execute('''select * from screen5_detail;''')
rows = cursor.fetchall()
df = pd.DataFrame(rows, columns=[col[0] for col in cursor.description])
df.head()
```

```python
df = df[['report_date', 'symbol', 'latest_price', '52wk_high','ema_9',
       'ema_21', 'adx_14', 'rsi_14', '2_week_rpi', '3_month_rpi',
       '6_month_rpi', '2wk_sma_of_6m_rpi']].sort_values(by=['symbol', 'report_date'], ascending=True).reset_index(drop=True)
```

```python
cursor = conn.cursor()
cursor.execute('''select * from screen5_trend;''')
rows = cursor.fetchall()
trend = pd.DataFrame(rows, columns=[col[0] for col in cursor.description])
trend.head()
```

```python
trend = trend[['report_date', 'symbol', '52wk_high',
       'trending_days', 'weighted_rpi', 'adx_14',
       'rsi_14']].sort_values(by=['symbol', 'report_date'], ascending=True).reset_index(drop=True)
```

```python
# join trend with df to get the latest price in trend as well 
trend = trend.merge(df[['report_date', 'symbol', 'latest_price']], on=['report_date', 'symbol'], how='left')
```

```python
import logging

# for each symbol and report date get the stock price from the yfinance API
def get_stock_prices(symbol, report_date):
    import yfinance as yf
    try:
        stock = yf.Ticker(symbol + '.NS')
        hist = stock.history(start=report_date)
        hist.reset_index(inplace=True)
        hist['symbol'] = symbol
        return hist[['Date', 'symbol', 'Close']].rename(columns={'Date': 'report_date', 'Close': 'price'})
    except Exception as e:
        logging.error(f"Error fetching data for {symbol}: {e}")
        return None
```

```python
s = get_stock_prices('KRISHANA', '2023-03-01')
```
